```python
#numpy task
import numpy as np
```

```python
#create numpy multidimentional array
arr = np.array([[10,20,30],
                [40,50,60],
                [70,80,90]])

print(arr)
```

```python
# accessing element
print("first Value =", arr[0,0])
print("second Row Third Value =", arr[1,2])
print("third Row Second Value =", arr[2,1])
```

```python
#addition
print(arr + 10)
```

```python
#substraction
print(arr - 5)
```

```python
#multiplication
print(arr * 2)
```

```python
#division
print(arr / 10)
```

```python
#mumpy methods

# 1. shape
print("1. Shape:", arr.shape)

# 2. size
print("2. Size:", arr.size)

# 3. ndim
print("3. Dimension:", arr.ndim)

# 4. dtype
print("4. Data Type:", arr.dtype)

# 5. max()
print("5. Maximum:", np.max(arr))

# 6. min()
print("6. Minimum:", np.min(arr))

# 7. sum()
print("7. Sum:", np.sum(arr))

# 8. mean()
print("8. Mean:", np.mean(arr))

# 9. reshape()
print("9. Reshape:\n", arr.reshape(1, 9))

# 10. flatten()
print("10. Flatten:", arr.flatten())

# 11. transpose()
print("11. Transpose:\n", arr.T)

# 12. sort()
print("12. Sort:\n", np.sort(arr))

# 13. sqrt()
print("13. Square Root:\n", np.sqrt(arr))

# 14. power()
print("14. Power:\n", np.power(arr, 2))

# 15. copy()
arr_copy = arr.copy()
print("15. Copy:\n", arr_copy)
```
