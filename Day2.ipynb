```python
# 1. LIST

# Creating a List
fruits = ["Apple", "Banana", "Mango"]

# append()
fruits.append("Orange")
print("Append:", fruits)

# insert()
fruits.insert(1, "Grapes")
print("Insert:", fruits)

# pop()
fruits.pop()
print("Pop:", fruits)

# remove()
fruits.remove("Banana")
print("Remove:", fruits)

# sort()
fruits.sort()
print("Sort:", fruits)
```

```python
# 2. DICTIONARY

# Creating a Dictionary
student = {
    "name": "Prashant",
    "age": 21,
    "city": "Surat"
}

# keys()
print("Keys:", student.keys())

# values()
print("Values:", student.values())

# items()
print("Items:", student.items())

# update()
student.update({"age": 22})
print("Update:", student)

# pop()
student.pop("city")
print("Pop:", student)
```

```python
# 3. TUPLE

# Creating a Tuple
numbers = (10, 20, 30, 20, 40,40,20)
print(numbers)

# count()
print("Count:", numbers.count(20))

# index()
print("Index:", numbers.index(30))

# len()
print("Length:", len(numbers))

# max()
print("Maximum:", max(numbers))

# min()
print("Minimum:", min(numbers))
```

```python
# SET
colors = {"Red", "Blue", "Green", "Red"}
print(colors)

# 1. add()
colors.add("Yellow")
print("After Add:", colors)

# 2. update() -
colors.update(["Black", "White"])
print("After Update:", colors)

# 3. remove()
colors.remove("Blue")
print("After Remove:", colors)

# 4. pop()
colors.pop()
print("After Pop:", colors)

# 5. clear()
colors.clear()
print("After Clear:", colors)
```

```python
# IF
age = 20
if age >= 18:
    print("Eligible to vote")
```

```python
# IF ELSE
age = 16

if age >= 18:
    print("Eligible")
else:
    print("Not Eligible")
```

```python
# IF ELIF ELSE
marks = 75

if marks >= 90:
    print("Grade A")
elif marks >= 60:
    print("Grade B")
else:
    print("Grade C")
```

```python
# NESTED IF
age = 20

if age >= 18:
    if age >= 21:
        print("adult")
    else:
        print("young adult")
else:
    print("minor")
```

```python
# BREAK
for i in range(1,6):
    if i == 5:
        break
    print(i)
```

```python
# CONTINUE
for i in range(1,6):
    if i == 3:
        continue
    print(i)
```

```python
# PASS
for i in range(1,6):
    if i == 3:
        pass
    print(i)
```

```python
# Take input from the user
name = input("Enter your name: ")
age = int(input("Enter your age: "))

# Display the input
print("Name:", name)
print("Age:", age)

if age >= 18:
    print("You are eligible to vote.")
else:
    print("You are not eligible to vote.")
```
